## Exploratory Data Analysis (EDA) and Aggregations

```python
import config
from clean import int_cols, float_cols, categorical_cols
```

```python
data_df = config.main()[1]
```

### Statistical Summaries of Columns

```python
for col_name in int_cols + float_cols:
    display(data_df[col_name].describe())
```

### Aggregations

Computing the average Energy Use Intensity (EUI) by Property Type.

```python
data_df.groupby(["Primary Property Type - Self Selected"])["Site EUI (GJ/m²)"].mean()
```

Computing the total Greenhouse Gas (GHG) emissions by year.

```python
print(data_df.groupby(["Primary Property Type - Self Selected","Year Ending"])["Total GHG Emissions (Metric Tons CO2e)"].sum().to_string())
```

```python
data_df.groupby(["Primary Property Type - Self Selected","Year Ending"])["Total GHG Emissions (Metric Tons CO2e)"].sum().unstack().plot(title="Total Greenhouse Gas (GHG) emissions by Property Type",kind="bar", rot=90)
```

Identifying the top 5 properties with the highest total energy consumption.

```python
data_df.groupby(["Property Name","Primary Property Type - Self Selected"])["Site Energy Use (GJ)"].mean().nlargest(5)
```

```python
data_df.groupby(["Property Name","Primary Property Type - Self Selected"])["Site Energy Use (GJ)"].mean().nlargest(5).plot.barh(title="Top 5 Properties by Energy Consumption", color="orange")
```
