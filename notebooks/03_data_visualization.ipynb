## Data Visualization

```python
import seaborn as sns
import config
```

```python
data_df = config.main()[1]
```

### Time-Series Visualization

Here, we plot the yearly trend of average Site Energy Use Intensity (EUI)

```python
grp = data_df.groupby(["Primary Property Type - Self Selected","Year Ending"])["Site EUI (GJ/m²)"].mean()
grp = grp.unstack(level=0)
grp
```

```python
plot = grp.plot(
                figsize=(10,18),
                kind="line",
                rot=90,
                grid=True,
                legend=True,
                title="Yearly Trend of Average Site Energy Use Intensity (EUI)",
            )
```

### Comparative Bar Charts

Below is a  bar chart showing the top 10 buildings with the highest GHG emissions. In each chart, each bar is annotated with emission values

```python
grp = data_df.groupby(["Property Name","Year Ending"])["Total GHG Emissions (Metric Tons CO2e)"].mean()
grp = grp.unstack(level=0).mean().nlargest(10)
grp
```

```python
plot = grp.plot(
                figsize=(12, 8),
                kind="bar",
                grid=True,
                legend=True,
                title="Top 10 Buildings with the Highest GHG Emissions"
            )

# Adding labels to the chart
plot.bar_label(plot.containers[0], label_type='edge')
plot.margins(y=0.1)
```

### Heatmap Visualization

The heatmap below shows the energy usage intensity (Site EUI (GJ/m²)) across different property 
types

```python
grp = data_df.groupby(["Primary Property Type - Self Selected","Year Ending"])["Site EUI (GJ/m²)"].mean()
grp = grp.unstack(level=0)
grp
```

```python
plot = sns.heatmap(grp, annot=True, fmt=".1f")
```
